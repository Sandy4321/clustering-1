```python
def make_table(itas_list):
    import numpy as np
    import pandas as pd
    
    itas = np.array(itas_list)
    
    n_papers = itas.shape[0]
    _titles = itas[list(range(0, n_papers, 2))]
    _authors = itas[list(range(1, n_papers, 2))]
    
    titles = []
    authors = []
    
    for title, author in zip(_titles, _authors):
        paper_authors = author.split(',')
        for _author in paper_authors:
            titles.append(title.strip())
            authors.append(_author.strip())
            
    return np.array(titles), np.array(authors)
```

```python
titles, authors = make_table(itas2009)
```

```python
import pandas as pd
import numpy as np
```

```python
titles = titles.reshape(192,1)
year = np.array([2009]*192).reshape(192,1)
authors = authors.reshape(192,1)
name = np.empty((192,1))
surname = np.empty((192,1))
author_id = np.empty((192,1))
gender = np.empty((192,1))
```

```python
authors[0][0]
```

```python
surname = []
for auth in authors:
    a = auth[0].split(' ')
    surname.append(a[0])
    
surname = np.array(surname).reshape(192,1)
```

```python
itas2009 = pd.DataFrame(data=np.concatenate([titles,
                                             year,
                                            authors,
                                            name,
                                            surname,
                                            author_id,
                                            gender], axis=1), columns=[['title','year','author',
                                                         'Name', 
                                                         'Surname',
                                                         'author_id_new',
                                                         'Gender']])
```
