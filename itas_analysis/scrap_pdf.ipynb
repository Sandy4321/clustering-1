```python
import urllib
from bs4 import BeautifulSoup
```

```python
def load_pdf_url(url):
    import urllib
    from bs4 import BeautifulSoup
    
    def process_pdf_url(url, main_url):
        
        main_url_list = main_url.split('.ru')
        
        if 'nopaper' in url:
            return np.nan
        else:
            url_list = url.split('pdf')
            if len(url_list) == 3:
                local_url = '/pdf' + url_list[1] + 'pdf'
            else:
                return np.nan
        return main_url_list[0] + '.ru' + local_url    
        
    
    response = urllib.request.urlopen(url)
    if response.status == 200:
        soup = BeautifulSoup(response, 'html.parser')
        
        titles = soup.find_all("a", {"class": "PaperTitle"})
        pdf = soup.find_all('a', onclick=True)

        titles_list = []
        pdf_list = []
        
        for title, pdf in zip(titles, pdf):
            titles_list.append(title.get_text())
            pdf_list.append(process_pdf_url(pdf['onclick'], url))        
            
        return titles_list, pdf_list
            
    else:
        print(response)
        return -1, -1
```

```python
itas2010 = 'http://www.itas2010.iitp.ru/search.html'
itas2011 = 'http://www.itas2011.iitp.ru/search.html'
itas2012 = 'http://www.itas2012.iitp.ru/search.html'
itas2013 = 'http://itas2013.iitp.ru/ru/search.html'
itas2014 = 'http://itas2014.iitp.ru/ru/search.html'
itas2015 = 'http://itas2015.iitp.ru/ru/search.html'
itas2016 = 'http://itas2016.iitp.ru/ru/search.html'
```

```python
titles, url = load_pdf_url(itas2010)

paper_url = pd.DataFrame(data={'titles':titles, 'url':url}, columns=[['titles', 'url']])


for itas_url in [itas2011, itas2012, itas2013, itas2014, itas2015, itas2016]:

    titles, url = load_pdf_url(itas_url)
    
    _itas = pd.DataFrame(data={'titles':titles, 'url':url}, columns=[['titles', 'url']])

    paper_url = pd.concat([paper_url, _itas])
```

```python
paper_url = paper_url.drop_duplicates('titles').dropna()
```
