```python
def parse_pdf_itas(path):
    
    import tika
    from tika import parser
    
    def decode_line(line):
        return line.encode('latin1').decode('cp1251').strip()
    
    def decode_char(line):
        _line = []
        for char in line:
            try:
                c = char.encode('latin1').decode('cp1251')
                c.append(_line)
            except:
                c = ''
        return ''.join(_line)
    
    
    pdf = parser.from_file(path)
    text_to_decode = pdf['content']
    text_to_decode = text_to_decode.strip().split('\n')
    
    text = []
    n_exceptions = 0
    for line in text_to_decode:
        try:
            single_line = decode_line(line)
            #single_line = line.encode('latin1').decode('cp1251').strip()
        except:
            n_exceptions += 1
            single_line = decode_char(line)
    
        text.append(single_line)
        
    return text
```

```python

def simple_parse(path):
    import tika
    from tika import parser
    pdf = parser.from_file(path)
    text_to_decode = pdf['content']
    text_to_decode = text_to_decode.strip().split('\n')
    
    return text_to_decode
```

```python
p = '/home/anvar/clustering/itas_analysis/pdf/'

_simple_parse = []
_complex_parse = []
for idx, path in enumerate(sorted(os.listdir(p))):
    
    try:
        text = parse_pdf_itas(p+path)
        text1 = simple_parse(p+path)
    except:
        print('Could not parse {}'.format(path))
        text = ['']
    if ('Список литературы' in text):
        print(idx, path, 'ru','parse_pdf_itas')
        _simple_parse.append(path)
    elif 'Список литературы' in text1:
        print(idx, path, 'ru','parse_itas')
        _complex_parse.append(path)
    elif 'References' in text1:
        print(idx, path, 'eng','parse_itas')
        _complex_parse.append(path)
    elif 'References' in text:
        print(idx, path, 'eng','parse_pdf_itas')
        _simple_parse.append(path)

        
    
```
