```python
def check_gender(name, surname):
    import re
    
    name = name.strip()
    surname = surname.strip()
    gender = np.nan

    if '.' not in name:
        if re.findall('[йцукенгшщзхъфывапролджэячсмитьбю]', name) != []:
            last_sym = name[-1]
            if (last_sym == 'а') or (last_sym == 'я') and (name not in ['Алипаша', 'Данила', 'Джошуа', 'Илья', 'Никита']): 
                gender = 'Female'
                return gender
            elif name == 'Аягоз':
                gender = 'Female'
                return gender
            else:
                gender = 'Male'
                return gender

    if np.isnan(gender):
        if re.findall('[йцукенгшщзхъфывапролджэячсмитьбю]', surname) != []:
            last_sym = surname[-1]
            if (last_sym == 'а') or (last_sym == 'я'):
                if surname in ['Ахуджа', 'Гайдамака', 'Дюба', 'Заика',
                               'Каледа', 'Калита', 'Катруца', 'Кикавада',
                               'Ксаба', 'Мажейка', 'Мостафа','Притула',
                               'Цепа','Ша','Шарануца', 'Шешкус','Ялда',]:
                    gender = np.nan
                else:
                    gender = 'Female'
                    return gender
            elif surname in ['Киселюс', 'Комеч', 'Корх', 'Крейдлин', 'Круглик',
                             'Крюк', 'Крячко', 'Лавыш', 'Лейн','Лик', 'Малько',
                             'Маляр', 'Матусевич','Мендель','Мовсесян','Молдован',
                             'Молибог','Мороз','Намиот','Пастушок','Пахл','Пилипчук',
                             'Пино','Поклад','Прун','Пэш','Птушенко','Рамзи','Раффлер',
                             'Ревенко','Ройтберг','Рябко','Рябых','Савчик','Саенко',
                             'Сергушевич','Стерлинг', 'Сухре', 'Тавакколках','Талис',
                              'Тамазян', 'Танас','Тетко','Ткачев', 'Фертиг','Фридман',
                              'Фришман', 'Фуденберг', 'Хамад', 'Хегази', 'Холод','Циммер',
                             'Чанг', 'Чаплин', 'Чевэн', 'Шинкарюк','Шмитт', 'Шнепс-Шнеппе',
                              'Эммерт-Стреиб']:
                gender = np.nan
            else:
                gender = 'Male'
                return gender
            
    return gender
```

```python
name = 'Пень.'
surname = 'Колода'

check_gender(name, surname)
```

```python
gender = []
for item in itas_data.iterrows():
    name = item[1].Name
    surname = item[1].Surname
    g = check_gender(name, surname)
    gender.append(g)
```

```python
itas_data['Gender'] = gender
```

```python
itas_data.sort_values(by=['year', 'author_id_new'], inplace=True)
itas_data.index = np.arange(2507)
```
